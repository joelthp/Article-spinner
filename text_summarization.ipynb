```python
import bs4 as bs
import urllib.request 
import re
import heapq
import nltk
from nltk.corpus import stopwords
```

```python
stopw=set(stopwords.words("english"))
```

```python
source=urllib.request.urlopen("https://en.wikipedia.org/wiki/Global_warming").read()
```

```python
soup=bs.BeautifulSoup(source,"html")
```

```python
paragraph=soup.findAll("p")
text=""
```

```python
for p in paragraph:
    text+=p.text
    
```

```python
text=re.sub(r"[\[0-9]*\]"," ",text)
```

```python
text=re.sub(r"[\[a-z]]"," ",text)
```

```python
text=re.sub(r"\[a-z]"," ",text)
```

```python
text
```

```python
clean_text=text.lower()
```

```python
clean_text=re.sub(r'\W',' ',clean_text)
```

```python
word2count={}
```

```python
for word in nltk.word_tokenize(clean_text):
    if word not in stopw:
        if word not in word2count.keys():
            word2count[word]=1
        else:
             word2count[word]+=1
for key in word2count.keys():
    word2count[key]/=max(word2count.values())           
    
```

```python
sent2score={}
sentences=nltk.sent_tokenize(text)
```

```python
for sent in sentences:
    for word in nltk.word_tokenize(sent):
        
        if word in word2count.keys():
            if len(sent.split(" "))<25:
                if word not in sent2score.keys():
                    sent2score[sent]=word2count[word]
                else:
                    sent2score[sent]+=word2count[word]
```

```python
best_sentence=heapq.nlargest(5,sent2score,key=sent2score.get)
```

```python
for sentence in best_sentence:
    print(sentence)
```
